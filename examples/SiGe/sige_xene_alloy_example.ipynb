```python
from ase import Atoms
from sqs import SQS
```

```python
a = 3.873
c = 30
buckling = 0.45879798303657/c
```

```python
silicene = Atoms('SiSi',scaled_positions=[[1/3,2/3,0.5+buckling/2],[2/3,1/3,0.5-buckling/2]],cell=[[a,0,0],[-0.5*a,3**0.5/2*a,0],[0,0,c]])
```

```python
sqs = SQS(max_m=4,conv_thr=[0,0,0,1e-3])
sqs.read_atoms(silicene,multiply=True,nx=8,ny=8,nz=1)
sqs.create_neighbor_list(twoD=True,verbose=True)
```

```python
silicene_sc = silicene.copy()

silicene_sc *= (8,8,1)
silicene_sc.write('POSCAR_0.vasp',vasp5=True,sort=True,direct=True)
```

```python
sqs.create_geometry(alloy_species='Ge',concentration=25,maxtrials=int(1e4),verbose=True)
sqs.write_sqs_atoms('POSCAR_25.vasp',vasp5=True,sort=True,direct=True)
sqs.save_state('state_25.json')
```

```python
sqs.create_geometry(alloy_species='Ge',concentration=50,maxtrials=int(1e4),verbose=True)
sqs.write_sqs_atoms('POSCAR_50.vasp',vasp5=True,sort=True,direct=True)
sqs.save_state('state_50.json')
```

```python
germanene = silicene.copy()
germanene.set_chemical_symbols(['Ge','Ge'])
germanene_sc = germanene.copy()

germanene_sc *= (8,8,1)
germanene_sc.write('POSCAR_100.vasp',vasp5=True,sort=True,direct=True)
```
